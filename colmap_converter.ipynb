```python
import pycolmap
import os

# Define input and output paths
# Ensure input_model_dir contains cameras.bin, images.bin, and points3D.bin
input_model_dir = '/home/isabelle/Projects/PhotoMesh' 
output_model_dir = '/home/isabelle/Projects/PhotoMesh/output'

# Create output directory if it doesn't exist
os.makedirs(output_model_dir, exist_ok=True)

try:
    # 1. Load the entire reconstruction from the binary directory
    reconstruction = pycolmap.Reconstruction(input_model_dir)
    
    print(f"Loaded reconstruction with {reconstruction.num_images()} images "
          f"and {reconstruction.num_points3D()} points.")

    # 2. Export to text format
    # PyCOLMAP writes all three files (cameras.txt, images.txt, points3D.txt) 
    # to the specified directory automatically.
    reconstruction.write_text(output_model_dir)

    print("Conversion complete! Files saved to:", output_model_dir)

except Exception as e:
    print(f"An error occurred during conversion: {e}")
```
