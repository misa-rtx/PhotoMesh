```python
import photomesh

result = photomesh.map_texture(
    dataset_dir="dataset/",
    mesh_name="poisson.ply",
    texture_size=4096,
    uv_method="xatlas",
    view_selector="closest_z",
    inpaint=True,
    color_match=True,            # match illumination across views
    color_match_method="mkl",
    color_match_ref=0,           # use first image as reference
    subdivide=0,                 # set to 2 for subdivision
    output_dir="outputs/",       # auto-save OBJ + PLY
)

mesh = result.get_open3d_mesh()
print(f"Open3D mesh: {len(mesh.vertices)} verts, {len(mesh.triangles)} tris")
```
